# Step 5: Outlier & Anomaly Detection

1. Boxplots for Outlier Detection (Numerical Features)

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(num_features, 1):
    plt.subplot(2,3,i)
    sns.boxplot(y=df[col], color="skyblue")
    plt.title(f"Outlier Detection in {col}")
plt.tight_layout()
plt.show()
```

2. Z-Score Method (Statistical Outliers)

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df[num_features]))
outliers = (z_scores > 3)  # threshold of 3 standard deviations
print("Number of outliers detected:", np.sum(outliers))

# Show rows with outliers
df_outliers = df[(outliers).any(axis=1)]
df_outliers.head()
```

3. IQR Method (Interquartile Range)

```python
def detect_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[col] < lower) | (data[col] > upper)]

for col in num_features:
    outliers_col = detect_outliers_iqr(df, col)
    print(f"{col}: {len(outliers_col)} outliers")
```

4. Visualizing Outliers with Target Variable

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(num_features, 1):
    plt.subplot(2,3,i)
    sns.boxplot(
        x='target',
        y=col,
        data=df,
        hue='target',
        palette='Set2',
        legend=False
    )
    plt.title(f"{col} by Heart Disease Status (Outliers Highlighted)")
plt.tight_layout()
plt.show()

```

# Step 6: PCA & Clustering

1. Data Preparation for PCA

```python
from sklearn.preprocessing import StandardScaler

# Select numerical features only (exclude target)
X = df[num_features]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

2. PCA (Dimensionality Reduction)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA results to dataframe
df['PCA1'] = X_pca[:,0]
df['PCA2'] = X_pca[:,1]

# Explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='target', data=df, palette='Set1', alpha=0.7)
plt.title("PCA of Heart Disease Dataset")
plt.show()
```

3. K-Means Clustering

```python
from sklearn.cluster import KMeans

# Apply KMeans with 2 clusters (disease vs no disease, but unsupervised)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Scatter plot with clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='Set2', alpha=0.7)
plt.title("KMeans Clustering on PCA Components")
plt.show()
```

4. Compare Clusters vs Actual Target

```python
pd.crosstab(df['Cluster'], df['target'])
```

5. Optional: Visualize Clusters by Risk

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='PCA1', y='PCA2',
    hue='Cluster',
    style='target',
    data=df,
    palette='Set2',
    alpha=0.7
)
plt.title("Cluster vs Actual Heart Disease Labels")
plt.show()
```

3. (Optional Advanced) Map Visualization with Plotly

```python
import plotly.express as px

# Dummy lat/long for cities
city_coords = {
    'Delhi': [28.6139, 77.2090],
    'Mumbai': [19.0760, 72.8777],
    'Bangalore': [12.9716, 77.5946]
}

# Map patient city to coordinates
df['lat'] = df['City'].map(lambda x: city_coords[x][0])
df['lon'] = df['City'].map(lambda x: city_coords[x][1])

# Aggregate by city
city_summary = df.groupby(['City','lat','lon'])['target'].mean().reset_index()

# Plot
fig = px.scatter_geo(
    city_summary,
    lat="lat", lon="lon",
    text="City",
    size="target",
    color="target",
    projection="natural earth",
    title="Average Heart Disease Prevalence by City (Dummy Data)"
)
fig.show()
```
